```python
max_dates = df.groupby('Customer ID')['InvoiceDate'].max()
max_dates = pd.to_datetime(max_dates, format='%d.%m.%Y %H:%M')
recency = (pd.to_datetime('01/01/2012', format='%d/%m/%Y') - max_dates).dt.days
```

```python
df['Recency'] = recency
```

```python
recency_revenue = df.groupby(['Recency'])['Revenue'].sum().reset_index()
recency_revenue['Revenue_Percentage'] = recency_revenue['Revenue'] / recency_revenue['Revenue'].sum() * 100
recency_revenue = recency_revenue.sort_values(by=['Recency'], ascending=False)
```

```python
df['Recency_Decile'] = pd.qcut(df['Recency'], q=10, labels=False)
```

```python
df['Recency_Decile']
```

```python
decile_revenue = df.groupby(['Recency_Decile'])['Revenue'].sum().reset_index()
```

```python
decile_revenue
```

```python
decile_revenue['Revenue_Percentage'] = decile_revenue['Revenue'] / decile_revenue['Revenue'].sum() * 100
```

```python
decile_revenue['Revenue_Percentage']
```

```python
decile_revenue = decile_revenue.sort_values(by=['Recency_Decile'])
```

```python
decile_revenue 
```

```python
cumulative_revenue = 0
for i in range(len(decile_revenue)):
    cumulative_revenue += decile_revenue.iloc[i]['Revenue_Percentage']
    if cumulative_revenue >= 75:
        lower_bound = decile_revenue.iloc[0]['Recency_Decile']
        upper_bound = decile_revenue.iloc[i]['Recency_Decile']
        print('Deciles', lower_bound, 'to', upper_bound, 'represent around 75% of the total revenue.')
        break
```
